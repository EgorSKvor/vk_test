```python
query_type_counts = df['query_type'].value_counts()
query_type_counts

# Визуализация
plt.figure(figsize=(8, 6))
sns.barplot(x=query_type_counts.index, y=query_type_counts.values)
plt.title('Распределение запросов по темам')
plt.xlabel('Тема запроса')
plt.ylabel('Количество запросов')
plt.show()
```

## doc_type = NaN, при event = search - это значит, что 4030 пользователей или 4677 раз во время поиска после введения запроса не сделали никаких действий - будем считать, что они неудовлетворены результатами поиска

```python
total_queries = df['query_id'].nunique()
actionable_queries = df[df['event'] != 'search']['query_id'].nunique()
efficiency = actionable_queries / total_queries * 100
print(f"Процент запросов без действий: {100-efficiency:.2f}%")
```

```python
# Типы запросов без действий
df[df['event'] == 'search']['query_type'].value_counts().head()
```

```python
# Запросы, где пользователи не выполняют никаких действий
problematic_queries = df[df['event'] == 'search']['query_text'].unique()
print("Проблемные запросы:")

# Топ-10 самых часто встречающихся проблемных запросов
top_problematic_queries = df[df['event'] == 'search']['query_text'].value_counts().head(10)
print("Топ-10 проблемных запросов:")

# Визуализация
plt.figure(figsize=(10, 6))
sns.barplot(x=top_problematic_queries.index, y=top_problematic_queries.values)
plt.title('Топ-10 проблемных запросов')
plt.xlabel('Текст запроса')
plt.ylabel('Количество раз')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
queries_per_user = df.groupby('user_id')['query_id'].nunique()
actions_per_query = df.groupby('query_id')['event'].count()

print(f"Среднее количество запросов на пользователя: {queries_per_user.mean():.2f}")
print(f"Среднее количество действий на запрос: {actions_per_query.mean():.2f}")
```

```python
print(f"Медиана количества запросов на пользователя: {queries_per_user.median()}")
print(f"Стандартное отклонение количества запросов на пользователя: {queries_per_user.std():.2f}")
```

## Это говорит о том, что количество запросов, которые делают пользователи, варьируется незначительно. В среднем 1 запрос на пользователя

```python
search_data = search.copy()
age_bins = [0, 18, 25, 35, 45, 55, 70]
age_labels = ['0-18', '18-24', '25-34', '35-44', '45-54', '55+']
search_data['age_group'] = pd.cut(search_data['age'], bins=age_bins, labels=age_labels)
age_group_distribution = search_data['age_group'].value_counts().sort_values(ascending=False)
age_group_distribution
```

```python
correlation_age_queries = search[['age', 'query_id']].corr().iloc[0, 1]
print(f"Корреляция между возрастом и количеством запросов: {correlation_age_queries:.2f}")
```

## Из анализа возрастных групп, неудовлетворенных поиском, можно сделать вывод, что по мере увеличения возраста кол-во запросов уменьшается, но из группировки по возрастным диапозонам видно, что чем старше пользователи, тем чаще они неудовлетворены поиском 

```python
search_data_by_age_query_type = search_data.groupby(['age_group','query_type'], observed=False)['user_id'].count()
unstacked = search_data_by_age_query_type.unstack(level='query_type')
plt.figure(figsize=(8, 6))
sns.heatmap(unstacked, annot=True, fmt='d', cmap='Blues')
plt.title('Тепловая карта количества запросов')
plt.xlabel('Тип запроса')
plt.ylabel('Возрастная группа')
plt.show()
```

```python
from collections import Counter
import re

# Разбиваем запросы на слова
all_words = ' '.join(search_data['query_text']).lower()
words = re.findall(r'\b\w+\b', all_words)
word_counts = Counter(words)

# Топ-20 самых популярных слов
top_words = word_counts.most_common(15)
top_words
```

## Из анализа тепловой карты и частотности слов в запросах делаем вывод, что мало неудовлетворенных пользователей, которые искали фильмы, в отличие от поиска новостей, шоу и музыки.
